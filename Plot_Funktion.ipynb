# Import 

```python
import numpy  as np 
import matplotlib.pyplot as plt

# Define a Function
def f(x):
    return x**2

# Create x and y Ranges
x = np.linspace(-10, 10, 10000)

y = f(x)
```

## Customise Plot

```python
plt.plot(x, y, 'r')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
