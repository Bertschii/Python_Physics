```python
from sympy import Matrix, symbols, cos, sin, diff, pi, init_printing

# Initialize pretty printing
init_printing(use_latex=True)

# Define symbols
omega, eta, t = symbols('omega eta t')

# Define the original matrix
matrix = Matrix([
    [cos(omega * t), -sin(omega * t) * cos(eta * t), sin(omega * t) * sin(eta * t)],
    [sin(omega * t), cos(omega * t) * cos(eta * t), -cos(omega * t) * sin(eta * t)],
    [0, sin(eta * t), cos(eta * t)]
])

# Compute the differential of the matrix with respect to t
differential_matrix = matrix.diff(t)

# Display the matrix and its differential in LaTeX form

print("Matrix")
display(matrix)

print("Differential der Matrix")
display(differential_matrix)

transpose_matrix = matrix.T

print("Transponierte Matrix")
display( transpose_matrix)
```

## Berchnen der schiefsymmetrische Matrix $\Omega$

```python
OmgeaProdukt =  differential_matrix * transpose_matrix

display(OmgeaProdukt)
```

Vereinfachen der Marix

```python
from sympy import simplify

simplify(OmgeaProdukt)
```
